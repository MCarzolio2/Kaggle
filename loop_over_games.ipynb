```python
import urllib2
from bs4 import BeautifulSoup
import game_level_parser as glp
from pymongo import MongoClient
```

```python
team_season_url = 'http://stats.ncaa.org/team/index/11540?org_id=164&sport_year_ctl_id=11540'
html_file = urllib2.urlopen(team_season_url)
soup = BeautifulSoup(html_file)


# Gives the list of urls for play-by-play data for team(season) given at team_season_url:
url_list = []

for a in soup.find_all('a', href=True):
    string = a['href']
    if string[1:5]=='game':
        game_id = string[12:19]
        url_list.append('http://stats.ncaa.org/game/play_by_play/'+game_id)
    
```
